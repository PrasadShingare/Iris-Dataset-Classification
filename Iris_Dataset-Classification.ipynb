# Training Machine Learning Model

## First Import Required Libraries 

```python
# importing alll the necessary packages to use the various classification algorithms
from sklearn.linear_model import LogisticRegression # for Logistic Regression Algorithm
from sklearn.model_selection import train_test_split # to split the dataset for training and testing 
from sklearn.neighbors import KNeighborsClassifier # KNN classifier
from sklearn import svm # for suport vector machine algorithm
from sklearn import metrics # for checking the model accuracy
from sklearn.tree import DecisionTreeClassifier # for using DTA
```

# Splitting The Data into Training And Testing Dataset

```python
X = iris[['sepal.length',	'sepal.width',	'petal.length',	'petal.width']].values
y = iris[['variety']]


X_train,X_test,y_train,y_test = train_test_split(X , y, test_size=0.2) # our main data split into train and test
# the attribute test_size=0.3 splits the data into 80% and 20% ratio. train=80% and test=20%

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Support Vector Machine SVM

```python
model = svm.SVC() # select the svm algorithm

# we train the algorithm with training data and training output
model.fit(X_train, y_train.values.ravel())

# we pass the testing data to the stored algorithm to predict the outcome
prediction = model.predict(X_test)
print('The accuracy of the SVM is: ', metrics.accuracy_score(y_test, prediction)) # we check the accuracy of the algorithm
#we pass the predicted output by the model and the actual output
```

# Logistic Regression

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train.values.ravel())
prediction = model.predict(X_test)
print('The accuracy of Logistic Regression is: ', metrics.accuracy_score(prediction, y_test)) #Check Accury
```

# Decision Tree

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
prediction = model.predict(X_test)
print('The accuracy of Decision Tree is: ', metrics.accuracy_score(prediction, y_test)) # Check Accuracy
```

# K-Nearest Neighbors

```python
model = KNeighborsClassifier(n_neighbors=3) # this examines 3 neighbors for putting the data into class
model.fit(X_train, y_train.values.ravel())
prediction = model.predict(X_test)
print('The accuracy of KNN is: ', metrics.accuracy_score(prediction, y_test)) #Check Accuracy 
```
